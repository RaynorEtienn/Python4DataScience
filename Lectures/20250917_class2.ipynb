```python
from collections import defaultdict

# needs to be a function (a callable at least)
def default_value():
    return 1

my_default = defaultdict(default_value)
```

```python
my_default[0]  # if a key is missing, inserts in the dict with value
# given by `default_value()`
```

```python
my_default
```

```python
list1 = ["mathurin", "badr", "julie"]
list2 = ["massias", "moufad", "alberge"]

for first in list1:
    print(first)
print()
# naive way to loop over 2 lists simultaneously:
for i in range(len(list1)):
    print(list1[i], list2[i])

print()
# pythonic way: use zip
for first, last in zip(list1, list2):
    print(first, last)
```

```python
# it stops at the shortest list:
list2 += ["lefebvre"]
for first, last in zip(list1, list2):
    print(first, last)

# you can zip as many lists as you want
```

```python
# how to iterate over both a list and its index?
# naive way: loop over the index and access the corresponding value
for i in range(len(list2)):
    print(f"Element number {i} is {list2[i]}")
```

```python
# the pythonic way : enumerate
for i, elem in enumerate(list2):
    print(f"Element number {i} is {elem}")

# conceptually, enumerate zips list2 with range(len(list2))
```

```python
# error workflow
my_dict = {12: "julie"}

# want to access a key, but I'm not sure it's there
# LBYL: look before you leap
key = 13
if key in my_dict:
    # maybe some other code before
    val = my_dict[key]
else:
    print("key not in dict")
```

```python
# EAFP easier to ask forgiveness than permission 
# try/except block:
try: # no permission part
    val = my_dict[key]  # throws an error
    print("Did I reach this part?")
except:  # ask for forgiveness part/ catches the error catching part
    print("key not in dict")
```

### "Advanced" for loops 

```python
# how do I check if `val` is present in `my_list` ?
```

```python
my_list = [chr(i) for i in range(52, 78)]  # nice `chr` function
print(my_list)
val = 'a'
found = False
for item in my_list:
    if item == val:
        found = True
print(found)
```

```python
val = '5'
found = False
for item in my_list:
    print(item)
    if item == val:
        found = True
        break  # stop the loop 
print(found)
```

```python
# the FOR/ELSE syntax:
val = 'a'
for item in my_list:
    if item == val:
        found = True
        print("Found")
        break  # stop the loop 
else:
    found = False
    print("Not found")
print(found)
```

```python
# the pass keyword
for i in range(12):
    if i % 2 == 0:
        print(i)
    else:
        pass  # placeholder to mean "do nothing" 
```

```python
# the continue keyword 
for i in range(12):
    if i % 2 == 0:
        # pass
        continue  # goes to the next iteration of the for loop
    if i % 3 == 0:
        print(i)
```
