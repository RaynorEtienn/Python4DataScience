````python
# FCwN 5

calculus = 1/10 + 100 - 100
print(f"{calculus = }")

print(f"{calculus == 0.1 = }")
````

### GPT 4.1 Answer:

The equality check for calculus == 0.1 is not working as expected due to floating-point precision errors in Python. When you compute 1/10 + 100 − 100, the result is not exactly 0.1 because 0.1 cannot be represented precisely in binary floating-point format. This leads to tiny rounding errors, so calculus is very close to 0.1, but not exactly equal, causing the comparison to return False.

````python
print(f"{np.isclose(calculus, 0.1) = }")
````

````python
# FCwB 8

import time

n = 10_000_000  # Large Number

python_begin = time.time()

python_product = 2
for k in range(1, n):
  python_product *= 4*k**2 / (4*k**2 - 1)

print(f"Python Time: {time.time() - python_begin:.3f} seconds")
print(f"{python_product = }")

numpy_begin = time.time()

numpy_product = 2 * np.prod(4 * np.square(np.arange(1, n)) / (4 * np.square(np.arange(1, n)) - 1))

print(f"Numpy Time: {time.time() - numpy_begin:.3f} seconds")
print(f"{numpy_product = }")
````

````python
# FCwB 8_b

%timeit 2 * np.prod(4 * np.square(np.arange(1, n)) / (4 * np.square(np.arange(1, n)) - 1))
````

### Our Answer:

```%timeit```: it is better because it benchmark properly the code by running it several time and giving back the mean and standard dev. of the code snippet.
```time.time```: it only execute code once, it is more customizable but in the end it only gives the time of computation.
